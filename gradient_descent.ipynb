```python
import numpy as np
import matplotlib.pyplot as plt

# Define the cost function (Mean Squared Error)
def cost_function(X, y, theta):
    m = len(y)  # Number of data points
    predictions = X.dot(theta)  # Predicted values
    cost = (1/(2*m)) * np.sum(np.square(predictions - y))  # MSE
    return cost

# Define the gradient descent function
def gradient_descent(X, y, theta, learning_rate, iterations):
    m = len(y)  # Number of data points
    cost_history = np.zeros(iterations)  # To store cost after each iteration
    
    for i in range(iterations):
        predictions = X.dot(theta)  # Predicted values
        theta = theta - (1/m) * learning_rate * (X.T.dot(predictions - y))  # Update theta
        cost_history[i] = cost_function(X, y, theta)  # Record the cost
        
    return theta, cost_history

# Data (input features and target values)
X = np.array([[1, 1], [1, 2], [1, 3], [1, 4]])  # Adding a column of 1s for the bias term
y = np.array([2, 2.5, 3.5, 4.5])

# Initial parameters (theta) and hyperparameters
theta = np.random.randn(2)  # Random initialization of theta
learning_rate = 0.01
iterations = 1000

# Run gradient descent
theta, cost_history = gradient_descent(X, y, theta, learning_rate, iterations)

# Output the results
print("Optimized parameters (theta):", theta)
print("Final cost:", cost_history[-1])

# Plot the cost function over iterations
plt.plot(range(iterations), cost_history, color='blue')
plt.title("Cost Function Over Iterations")
plt.xlabel("Number of Iterations")
plt.ylabel("Cost (MSE)")
plt.show()
```
